```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from sklearn.metrics import classification_report,confusion_matrix
```

<h3>Load data</h3>

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

```python
print("Training data shape:", x_train.shape)  # Shape of training images
print("Training labels shape:", y_train.shape)  # Shape of training labels
print("Test data shape:", x_test.shape)  # Shape of test images
print("Test labels shape:", y_test.shape)  # Shape of test labels
```

```python
classes = ["airplane","automobile","bird","cat","deer","dog","frog","horse","ship","truck"]
```

```python
def plot_image(X,y,i):
    plt.figure(figsize = (15,2))
    plt.imshow(X[i])
    plt.title(classes[y[i][0]])
```

```python
plot_image(x_train,y_train,33)
```

```python
plot_image(x_train,y_train,233)
```

<h3>Normalize data</h3>

```python
x_train_normalized=x_train/255
x_test_normalized=x_test/255
```

<h3> categorical y </h3>

```python
y_train_categorical = to_categorical(y_train, num_classes=10)
y_test_categorical = to_categorical(y_test, num_classes=10)
```

<h3>Build CNN model </h3>

```python
cnn=Sequential([
    Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    
    Flatten(),
    Dense(64,activation='relu'),
    Dense(32,activation='relu'),
    Dense(10,activation='softmax'),
])
cnn.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
cnn.fit(x_train_normalized,y_train_categorical)
```

```python
cnn.evaluate(x_test_normalized,y_test_categorical)
```

```python
y_pred=cnn.predict(x_test_normalized)
for i,y in enumerate(y_pred):
    arg=np.argmax(y)
    for j in range(10):
        y_pred[i][j]=0
        if arg==j:
            y_pred[i][j]=1
```

```python
print(classification_report(y_test_categorical,y_pred))
```

```python
cm = confusion_matrix(y_test_categorical.argmax(axis=1), y_pred.argmax(axis=1))

# Display the confusion matrix using Seaborn's heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```
